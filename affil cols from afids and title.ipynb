```python
MY_API_KEY='592fce417ee8847ba636e275613f4cc6'
count = 0
for index,row in df.iterrows():
    count +=1
    print(count)
    print(type(row["affid"]), row["affid"])
    try:
        aff = AffiliationRetrieval(row["affid"])
        df.at[index, 'affiliation_name'] = aff.affiliation_name
        df.at[index, 'affiliation_country'] = aff.country
        df.at[index, 'affiliation_city '] = aff.city
        df.at[index, 'affiliation_url'] = aff.url
        df.at[index, 'affiliation_org_url'] = aff.org_URL

        print(aff.affiliation_name)
        print(aff.city)
        print(aff.country)
        print(aff.org_URL)
        print(aff.url)
        print()
    except:
        try:
            print("from AF_ID(query)")                            
            query = "AF-ID(" + str(row["affid"]) + ")"
            s = AffiliationSearch(query)

            df.at[index, 'affiliation_name'] = s.affiliations[0].name
            df.at[index, 'affiliation_country'] = s.affiliations[0].country
            df.at[index, 'affiliation_city '] = s.affiliations[0].city
            df.at[index, 'affiliation_url'] = "https://api.elsevier.com/content/affiliation/affiliation_id/" + str(row["affid"])
            df.at[index, 'affiliation_org_url'] = None

            print(df.at[index, 'affiliation_name'])
            print(df.at[index, 'affiliation_city'])
            print(df.at[index, 'affiliation_country'])
            print(df.at[index, 'affiliation_url'])
            print(df.at[index, 'affiliation_org_url'])
            print()
            
        except:
            print("from title")
            title = row["title"].replace("(","").replace(")","")
            resp = requests.get("https://api.elsevier.com/content/search/scopus?query=( TITLE ("+ title +" ) )",
                        headers={'Accept':'application/json',
                                 'X-ELS-APIKey': MY_API_KEY})
            results = resp.json()
            entries = results['search-results']['entry']
            for i in range(len(entries)):
                if 'affiliation' in entries[i]:
                    for j in range(len( entries[i]['affiliation'])):
                        if 'afid' in entries[i]['affiliation'][j] and entries[i]['affiliation'][j]['afid'] is not None:
                            if(row["affid"] == entries[i]['affiliation'][j]['afid']):

                                if 'affilname' in entries[i]['affiliation'][j] and entries[i]['affiliation'][j]['affilname'] is not None:
                                    df.at[index, 'affiliation_name'] = entries[i]['affiliation'][j]['affilname']
                                else:
                                    df.at[index, 'affiliation_name'] = None

                                if 'affiliation-city' in entries[i]['affiliation'][j] and entries[i]['affiliation'][j]['affiliation-city'] is not None:
                                    df.at[index, 'affiliation_city'] = entries[i]['affiliation'][j]['affiliation-city']
                                else:
                                    df.at[index, 'affiliation_city'] = None                    

                                if 'affiliation-country' in entries[i]['affiliation'][j] and entries[i]['affiliation'][j]['affiliation-country'] is not None:
                                    df.at[index, 'affiliation_country'] = entries[i]['affiliation'][j]['affiliation-country']
                                else:
                                    df.at[index, 'affiliation_country'] = None

                                if 'affiliation-url' in entries[i]['affiliation'][j] and entries[i]['affiliation'][j]['affiliation-url'] is not None:
                                    df.at[index, 'affiliation_url'] = entries[i]['affiliation'][j]['affiliation-url']
                                else:
                                    df.at[index, 'affiliation_url'] = None

                                if 'affiliation-org-url' in entries[i]['affiliation'][j] and entries[i]['affiliation'][j]['affiliation-org-url'] is not None:
                                    df.at[index, 'affiliation_org_url'] = entries[i]['affiliation'][j]['affiliation-url']
                                else:
                                    df.at[index, 'affiliation_org_url'] = None
           
```
