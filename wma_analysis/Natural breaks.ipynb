```python
import pandas as pd
import jenkspy
from matplotlib import pyplot as plt
from jenkspy import JenksNaturalBreaks
```

```python
df = pd.read_excel('WMA pneumonia cocci with weather data_annual avg reset.xlsx')
```

```python
df
```

```python
df_new = df[df['Average of Temperature mean (⁰F)'].notnull()]
```

```python
df_new
```

```python
df_new = df_new[df_new['WMA Positivity Rate(variable days)'].notnull()]
```

```python
df_new
```

```python
jnb = JenksNaturalBreaks(5)
```

```python
jnb.fit(df_new['WMA Positivity Rate(variable days)'].tolist())
```

```python
print(jnb.labels_) # Labels for fitted data
print(jnb.groups_) # Content of each group
print(jnb.breaks_) # Break values (including min and max)
print(jnb.inner_breaks_)
```

```python
breaks = jenkspy.jenks_breaks(df_new['WMA Positivity Rate(variable days)'], n_classes=5)
print(breaks)
```

```python
df_new['positivity_rates_category'] = pd.cut(df_new['WMA Positivity Rate(variable days)'], bins=breaks, labels=['category_1', 'category_2','category_3', 'category_4','category_5'], include_lowest=True)
```

```python
df_new
```

```python
df_new.plot.scatter(x='WMA Positivity Rate(variable days)', y = 'positivity_rates_category')
```

```python
df_new.to_excel('WMA positivity_rates_categories.xlsx', index=False)
```
