```python
import pandas as pd
import jenkspy
from matplotlib import pyplot as plt
from jenkspy import JenksNaturalBreaks
```

```python
df = pd.read_excel('WMA pneumonia cocci with weather data_annual avg reset.xlsx')
```

```python
df
```

```python
df_new = df[df['Average of Temperature mean (‚Å∞F)'].notnull()]
```

```python
df_new
```

```python
df_new = df_new[df_new['WMA Positivity Rate(variable days)'].notnull()]
```

```python
df_new
```

Lead and Lag

```python
df_new['30 day avg. humidity_lag'] = df_new['30 day avg. humidity'].shift(20)
```

```python
df_new['30 day avg. temp_lead'] = df_new['30 day avg. temp'].shift(-43)
```

```python
df_new['30 day avg. wind_lag'] = df_new['30 day avg. wind'].shift(52)
```

```python
df_new
```

```python
df_new = df_new[df_new['30 day avg. wind_lag'].notnull() & df_new['30 day avg. temp_lead'].notnull() & df_new['30 day avg. humidity_lag'].notnull()]
```

```python
df_new
```

Date filter

```python
df_new['Date']
```

```python
df_new = df_new[(df_new['Date'].dt.year == 2020) | (df_new['Date'].dt.year == 2021) | (df_new['Date'].dt.year == 2022)]
df_new
```

```python
df_new.to_excel('WMA positivity_rates_categories_2020_2021_2022_leads_lags.xlsx', index=False)
```

Natural breaks
5 categories

```python
jnb = JenksNaturalBreaks(5)
```

```python
jnb.fit(df_new['WMA Positivity Rate(variable days)'].tolist())
```

```python
print(jnb.labels_) # Labels for fitted data
print(jnb.groups_) # Content of each group
print(jnb.breaks_) # Break values (including min and max)
print(jnb.inner_breaks_)
```

```python
breaks = jenkspy.jenks_breaks(df_new['WMA Positivity Rate(variable days)'], n_classes=5)
print(breaks)
```

```python
df_new['positivity_rates_5_categories'] = pd.cut(df_new['WMA Positivity Rate(variable days)'], bins=breaks, labels=['category_1', 'category_2','category_3', 'category_4','category_5'], include_lowest=True)
```

```python
df_new
```

```python
df_new.plot.scatter(x='WMA Positivity Rate(variable days)', y = 'positivity_rates_5_categories')
```
