```python
import pandas as pd
import numpy as np
```

```python
df_original = pd.read_excel("scopus_master_ref_formatted2_aff_names_dctitle_y_dup_removed_230614.xlsx")
```

```python
df = df_original[["eid","STUDY TITLE/NAME","affiliation-url"]]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df
```

```python
df_new = pd.DataFrame()
eid = []
title = []
affid = []
df_new['eid'] = eid
df_new['title'] = title
df_new['affid'] = affid
```

```python
df_new
```

```python
url_text = "https://api.elsevier.com/content/affiliation/affiliation_id/"
affid_original_sum=0
for index, row in df.iterrows():
    url_list = []
    url_list = row["affiliation-url"].replace(url_text, "").split(",")
    
    for i in url_list:
        df_new.loc[len(df_new)] = [row["eid"], row["STUDY TITLE/NAME"],i]
        
    affid_original_sum+=len(url_list)
    print(len(url_list))
    print(url_list)
```
