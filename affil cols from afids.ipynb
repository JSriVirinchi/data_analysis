```python
df['affiliation_name'] = ""
df['affiliation_country'] = ""
df['affiliation_city'] = ""
df['affiliation_url'] = ""
df['affiliation_org_url'] = ""
```

```python
df = df.drop_duplicates(subset=['affid'], keep='first')
```

```python
df.iloc[11]
```

```python
count = 0
for index,row in df.iterrows():
    count +=1
    print(count)
    print(type(row["affid"]), row["affid"])
    try:
        aff = AffiliationRetrieval(row["affid"])
    except:
        print("error")
        continue
        
    df.at[index, 'affiliation_name'] = aff.affiliation_name
    df.at[index, 'affiliation_country'] = aff.country
    df.at[index, 'affiliation_city '] = aff.city
    df.at[index, 'affiliation_url'] = aff.url
    df.at[index, 'affiliation_org_url'] = aff.org_URL

    print(aff.affiliation_name)
    print(aff.sort_name)
    print(aff.org_type)
    print(aff.postal_code)
    print(aff.city)
    print(aff.state)
    print(aff.country)
    print(aff.org_domain)
    print(aff.org_URL)
    print(aff.url)
    print()
    
```
